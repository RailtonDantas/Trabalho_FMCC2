```python
# imports

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Lendo dados
des = pd.read_csv("../resultado_DES.csv")
des["algoritmo"] = "DES"

rsa = pd.read_csv("../resultado_RSA.csv")
rsa["algoritmo"] = "RSA"

gm = pd.read_csv("../resultado_Goldwasser-micali.csv")
gm["algoritmo"] = "Goldwasser-Micali"

df = pd.concat([des, rsa, gm], ignore_index=True)
```

```python
sns.set_theme(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df,
    x="Tamanho_msg_MB",
    y="Tempo_gasto_ms",
    hue="algoritmo",
    marker="o"
)

plt.title("Comparação de Tempo de Execução")
plt.xlabel("Tamanho da Mensagem (MB)")
plt.ylabel("Tempo Gasto (ms)")
plt.legend(title="Algoritmo")
plt.grid(True)
plt.show()
```
